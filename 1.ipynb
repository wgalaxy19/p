```python
#simulation and display of an image 


from PIL import Image,ImageOps

def print_image(image_path):
    img = Image.open(image_path)
    display(img)

    img_negative = ImageOps.invert(img)
    display(img_negative)

    gray_scale= ImageOps.grayscale(img)
    display(gray_scale)

    gray_scale_inverted = ImageOps.invert(gray_scale)
    display(gray_scale_inverted)

    binary_image = gray_scale.point(lambda p: p > 128 and 255)
    display(binary_image)

    binary_negative = ImageOps.invert(binary_image)
    display(binary_negative)

image_path="flow.jpeg"
print_image(image_path)
```
