```python
#Image intensity slicing technique for image enhancement
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("flow.jpeg",0)
row, column = img.shape
img1 = np.zeros((row, column), dtype='uint8')
min_range = 10
max_range = 50
for i in range(row):
    for j in range(column):
        if min_range < img[i, j] < max_range:
            img1[i, j] = 255
        else:
            img1[i, j] = 0

plt.figure(figsize=(12,6))
plt.subplot(241)
plt.title("Original Image")
plt.imshow(img,cmap="gray")
plt.axis("off")
plt.subplot(242)
plt.title("Sliced Image")
plt.imshow(img1,cmap="gray")
plt.axis("off")
plt.tight_layout()
plt.show()
```
