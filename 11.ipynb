```python
import cv2
import numpy as  np
import matplotlib.pyplot as plt

img = cv2.imread("flow.jpeg")

blur_image = cv2.GaussianBlur(img,(3,3),0)
edges = cv2.Canny(image=blur_image,threshold1=100,threshold2=200)

plt.figure(figsize=(12,8))

plt.subplot(131)
plt.title("Origianal image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(132)
plt.title("Blurred image")
plt.imshow(cv2.cvtColor(blur_image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(133)
plt.title("Canny Edge of an  image")
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.tight_layout()
plt.show()
```
