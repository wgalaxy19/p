```python
#7.Computation of Mean,Standard Deviation,Correlation coefficient of the given image
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

im1= Image.open("flower.jpeg")
im2= Image.open("car.jpeg")

plt.figure(figsize=(12,6))

plt.subplot(221)
plt.title("First Image Before Resize")
plt.imshow(im1)
plt.axis("off")

plt.subplot(222)
plt.title("Second Image Before Resize")
plt.imshow(im2)
plt.axis("off")

img1 = np.array(im1)
img2 = np.array(im2)
img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))

plt.subplot(223)
plt.title("First Image After Resize")
plt.imshow(img1)
plt.axis("off")

plt.subplot(224)
plt.title("Second Image After Resize")
plt.imshow(img2)
plt.axis("off")

mean_img1 = np.mean(img1)
std_img1 = np.std(img1)
mean_img2 = np.mean(img2)
std_img2 = np.std(img2)
corr_coeff = np.corrcoef(img1.flatten(), img2.flatten())[0, 1]
print("Mean of first image : ", mean_img1)
print("Mean of second image : ", mean_img2)
print("Standard Deviation of first image : ", std_img1)
print("Standard Deviation of second image : ", std_img2)
print("Correlation Coefficient : ", corr_coeff)
```
